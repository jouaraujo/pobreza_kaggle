```python
import pandas as pd
import matplotlib.pyplot as plt;
```

```python
df = pd.read_csv("data/indices_pobreza_consolidado.csv")
df
```

```python
df["referencia"] = pd.to_datetime(df["referencia"])
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, gridspec_kw={'width_ratios': [2, 2]}, figsize=(12, 6))
ax1.set_title("Pobreza no período de 10 anos(2012 à 2022)")
ax1.title.set_position([.5, 1.05])
ax1.plot(df["referencia"], df["pobreza"], c="#ad775c")
ax1.grid()
ax1.set_ylabel("Taxa de pobreza")
ax1.set_xlabel("Anos")

ax2.set_title("Extrema-Pobreza no período de 10 anos(2012 à 2022)")
ax2.title.set_position([.5, 1.05])
ax2.plot(df["referencia"], df["extrema_pobreza"], c="#47775c")
ax2.grid()
ax2.set_ylabel("Taxa de extrema_pobreza")
ax2.set_xlabel("Anos");
```
