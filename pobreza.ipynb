```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt;
```

```python
df = pd.read_csv("data/indices_pobreza_consolidado_anual.csv")
df
```

```python
df["periodo"] = pd.to_datetime(df["periodo"])
```

```python
df.dtypes
```

```python
df.shape
```

```python
df.columns
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, gridspec_kw={'width_ratios': [2, 2]}, figsize=(12, 6))
ax1.set_title("Pobreza no período de 10 anos(2012 à 2022)")
ax1.title.set_position([.5, 1.05])
ax1.plot(df["referencia"], df["pobreza"], c="green")
ax1.grid()
ax1.set_ylabel("Taxa de pobreza")
ax1.set_xlabel("Anos")

ax2.set_title("Extrema-Pobreza no período de 10 anos(2012 à 2022)")
ax2.title.set_position([.5, 1.05])
ax2.plot(df["referencia"], df["extrema_pobreza"], c="yellow")
ax2.grid()
ax2.set_ylabel("Taxa de extrema_pobreza")
ax2.set_xlabel("Anos");
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, gridspec_kw={'width_ratios': [2, 2]}, figsize=(12, 6))
ax1.set_title("familias pobres no período de 10 anos(2012 à 2022)")
ax1.title.set_position([.5, 1.05])
ax1.plot(df["referencia"], df["familias_pobreza"], c="blue")
ax1.grid()
ax1.set_ylabel("Taxa de familias pobreza")
ax1.set_xlabel("Anos")

ax2.set_title("Familias extrema-pobreza no período de 10 anos(2012 à 2022)")
ax2.title.set_position([.5, 1.05])
ax2.plot(df["referencia"], df["familias_extrema_pobreza"], c="purple")
ax2.grid()
ax2.set_ylabel("Taxa de familias extrema pobreza")
ax2.set_xlabel("Anos");
```
