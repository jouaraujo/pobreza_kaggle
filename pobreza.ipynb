```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt;
```

```python
df = pd.read_csv("indices_pobreza_consolidado_anual.csv")
df
```

```python
df["periodo"] = pd.to_datetime(df["periodo"])
```

```python
plt.plot(df["referencia"], df["pobreza"], c="green")
plt.title("Pobreza no período de 10 anos(2012 à 2022)")
plt.grid()
plt.ylabel("Taxa de pobreza")
plt.xlabel("Anos");
```

```python
plt.plot(df["referencia"], df["extrema_pobreza"], c="yellow")
plt.title("Extrema_pobreza no período de 10 anos(2012 à 2022)")
plt.grid()
plt.ylabel("Taxa de extrema_pobreza")
plt.xlabel("Anos");
```
