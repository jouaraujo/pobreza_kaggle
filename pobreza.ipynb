```python
import pandas as pd
import matplotlib.pyplot as plt

pd.set_option('display.max_columns', 200)
plt.style.use('seaborn-v0_8-whitegrid')
```

# Carregando os dados 

primeiro dataframe indices_pobreza_consolidado

```python
df_consolidado = pd.read_csv("data/indices_pobreza_consolidado.csv")
df_consolidado.head()
```

```python
# informações à respeito do df_consolidado
df_consolidado.info()
```

segundo dataframe df_consolidado_anual

```python
df_consolidado_anual = pd.read_csv("data/indices_pobreza_consolidado_anual.csv")
df_consolidado_anual.head()
```

```python
# informações à respeito do df_consolidado
df_consolidado_anual.info()
```

```python
# convertendo a coluna periodo do df_consolidade para o formato data
df_consolidado.periodo = pd.to_datetime(df_consolidado.periodo, format='%Y%m')
```

# Visualização 

### Gráfico de linhas

```python
# Gráfico de linha comparando pobreza e extrema pobreza (2012 à 2022)
fig, (ax1, ax2) = plt.subplots(1, 2,  figsize=(15, 5))
ax1.set_title("Pobreza no período de 10 anos(2012 à 2022)")
ax1.plot(df_consolidado["periodo"], df_consolidado["pobreza"])
ax1.set_ylabel("Taxa de pobreza")
ax1.set_xlabel("Anos")

ax2.set_title("Extrema-Pobreza no período de 10 anos(2012 à 2022)")
ax2.plot(df_consolidado["periodo"], df_consolidado["extrema_pobreza"])
ax2.set_ylabel("Taxa de extrema_pobreza")
ax2.set_xlabel("Anos")
plt.show()
```

```python
# Gráfico de linha comparando familias pobres e familias em extrema-pobreza (2012 à 2022)
fig, (ax1, ax2) = plt.subplots(1, 2,  figsize=(15, 5))
ax1.set_title("familias pobres no período de 10 anos(2012 à 2022)")
ax1.plot(df_consolidado["periodo"], df_consolidado["familias_pobreza"])
ax1.set_ylabel("Taxa de familias pobreza")
ax1.set_xlabel("Anos")

ax2.set_title("Familias extrema-pobreza no período de 10 anos(2012 à 2022)")
ax2.plot(df_consolidado["periodo"], df_consolidado["familias_extrema_pobreza"])
ax2.set_ylabel("Taxa de familias extrema pobreza")
ax2.set_xlabel("Anos")
plt.show()
```

### Gráfico boxplots

```python
# Gráfico de linha comparando indigenas pobres e indigenas extremamente-pobres (2012 à 2022)
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))
ax1.set_title("Indigenas pobreza no período de 10 anos(2012 à 2022)")
ax1.plot(df_consolidado["periodo"], df_consolidado["indigenas_pobreza"])
ax1.set_ylabel("Taxa de indigenas pobreza")
ax1.set_xlabel("Anos")

ax2.set_title("indigenas extrema-Pobreza no período de 10 anos(2012 à 2022)")
ax2.plot(df_consolidado["periodo"], df_consolidado["indigenas_extrema_pobreza"])
ax2.set_ylabel("Taxa de indigenas extrema pobreza")
ax2.set_xlabel("Anos");
```

```python
# Boxplot de quilombolas pobres e extremamento pobres (2012 à 2022)
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))
ax1.set_title("Quilom. pobreza no período de 10 anos(2012 à 2022)")
ax1.boxplot(df_consolidado["quilombolas_pobreza"])
ax1.set_ylabel("Taxa de quilombolas pobreza")
ax1.plot(1, df_consolidado.quilombolas_pobreza.mean(), 'ro', label='média')
ax1.legend()

ax2.set_title("Quilom. extrema-Pobreza no período de 10 anos(2012 à 2022)")
ax2.boxplot(df_consolidado["quilombolas_extrema_pobreza"])
ax2.set_ylabel("Taxa de quilombolas extrema pobreza")
ax2.plot(1, df_consolidado.quilombolas_extrema_pobreza.mean(), 'ro', label='média')
ax2.legend()
plt.show()
```

```python
# Boxplot de ciganos pobres e extremamento pobres (2012 à 2022)
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))
ax1.set_title("Ciganos pobreza no período de 10 anos(2012 à 2022)")
ax1.boxplot(df_consolidado["ciganos_pobreza"])
ax1.set_ylabel("Taxa de ciganos pobreza")
ax1.plot(1, df_consolidado.ciganos_pobreza.mean(), 'ro', label='média')
ax1.legend()

ax2.set_title("Ciganos extrema-Pobreza no período de 10 anos(2012 à 2022)")
ax2.boxplot(df_consolidado["ciganos_extrema_pobreza"])
ax2.set_ylabel("Taxa de ciganos extrema pobreza")
ax2.plot(1, df_consolidado.ciganos_extrema_pobreza.mean(), 'ro', label='média')
ax2.legend()
plt.show()
```
